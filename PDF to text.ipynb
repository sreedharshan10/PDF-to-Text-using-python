```python
import PyPDF2
```

```python
# Open the PDF file in binary mode
pdf = open("Sreeee.pdf", "rb")

# Create a PdfFileReader object from the binary file object
reader = PyPDF2.PdfReader(pdf)

# Get the first page of the PDF
page = reader.pages[0]

# Extract the text from the page using the extractText() method
text = page.extract_text()

# Print the extracted text
print(text)

# Close the PDF file
pdf.close()
```
